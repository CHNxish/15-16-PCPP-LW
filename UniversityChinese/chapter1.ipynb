## 寡人之于国也

  - 节选《孟子·梁惠王上》
  
  - 四书：《大学》 《论语》 《中庸》 《孟子》
  
  - 寡人：寡德之人，道德方面做得不足的人，古代君王的谦称。
  
  - 课文结构：提出民不加多的疑问，分析民不加多的原因，论述民加多的途径。线索：民不加多。
  
```
梁惠王曰：“寡人之于国也，尽心焉耳矣。河内凶，则移其民于河东，移其粟于河内；河东凶亦然。察邻国之政，无如寡人之用心者。邻国之民不加少，寡人之民不加多，何也？”

# 凶：灾荒

孟子对曰：“王好战，请以战喻。填然鼓之，兵刃既接，弃甲曳兵而走。或百步而后止，或五十步而后止。以五十步笑百步，则何如？”

# 填：拟声词 接：交锋 

梁惠王曰：“不可，直不百步耳，是亦走也。”

# 直：只，仅

孟子曰：“王如知此，则无望民之多于邻国也。”

“不违农时，谷不可胜食也；数罟不入洿池，鱼鳖不可胜食也；斧斤以时入山林，材木不可胜用也。谷与鱼鳖不可胜食，材木不可胜用，是使民养生丧死无憾也。养生丧死无憾，王道之始也。”

“五亩之宅，树之以桑，五十者可以衣帛矣；鸡豚狗彘之畜，无失其时，七十者可以食肉矣；百亩之田，勿夺其时，数口之家可以无饥矣；谨庠序之教，申之以孝悌之义，颁白者不负戴于道路矣。七十者衣帛食肉，黎民不饥不寒，然而不王者，未之有也。”

“狗彘食人食而不知检，涂有饿莩而不知发，人死，则曰：‘非我也，岁也。’是何异于刺人而杀之，曰：‘非我也，兵也。’？王无罪岁，斯天下之民至焉。”
```

  - 问题：概括本文所体现的王道政治的主要内容，体会孟子以民为本的主要思想如何体现。
  
    - 孟子首先指出“养生丧死无憾，王道之始也。”国君只要做到保障农时、谷物、材木等基本生产，使百姓的基本生活得以保障，才算是王道政治的开始。
    
    - 孟子接着指出“七十者衣帛食肉，黎民不饥不寒，然而不王者，未之有也。”国君若能进一步提高百姓的生活水平，使百姓达到无饥、衣帛、食肉，在此基础上，继续发展教育，改善民风，使社会形成尊老互助的良好风气，那么王道政治便得以实现了。
    
    - 孟子最后运用比喻示意梁惠王，国君切不可推卸责任。
    
    - 综上，孟子的建议均围绕百姓为中心展开论述，体现了“民重君轻”的民本思想。
    
  - 问题：体会孟子运用比喻说理的特点。
  
    - 以作战比喻治国，通过“五十步笑百步”说明梁惠王之政与邻国之政并无本质区别。
    
    - 以杀人却推卸责任给兵器比喻君王推脱百姓饿死之责任给灾荒说明国君不能推卸责任。
    
  - 问题：体会孟子抑扬兼施、循循善诱的说理艺术。
  
    - 以作战比喻治国，指出梁惠王治国之政与别国无本质区别，都未实行仁政，然后提出实行王道的具体措施；以杀人比喻国君推卸责任，然后指出君王不归罪于收成不好便可使“民加多”。以上都是先抑后扬。
    
    - 在打消梁惠王矜傲的同时，又能抓住他渴望民众拥戴的潜在心理进行诱导；在阐述实行王道的具体措施时，从保障基本生活-提高生活水平-发展教育、改善民风，由易到难步步推进。以上为循循善诱的体现。
